```python
# хостинге MongoDB,
# записать собранные вакансии в созданную БД

# Импорт библиотек
from pymongo import MongoClient
import pandas as pd
from pprint import pprint

# Подключение к БД - клиент
client = MongoClient('localhost',27017)
# создание базы данных
db = client['db_vacancies']
# выбираем коллекцию документов
hh = db.hh




# Загрузка датафрейма вакансий  полученного в предыдущем упражнении

df_hh = pd.read_excel('https://drive.google.com/uc?id=1qESyDQPZDLbGGokaeTGeheyIEMGdBBKF')

df_hh.T

```

```python

# функция добавления в базу
def DF_to_DB(df, collection):
    for i in range(df.shape[0]):
        vac = df.loc[i]
        dictionary = {
            'name': vac['Имя'],
            'link': vac["Ссылка"],
            'salary_min': vac["Мин. зп"],
            'salary_max': vac['Макс. зп'],
            'source': vac["Источник"]}
        collection.insert_one(dictionary)

DF_to_DB(df_hh, hh)
```
